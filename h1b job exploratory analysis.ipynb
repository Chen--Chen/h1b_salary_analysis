```python
# Import libraries
import mysql.connector
import pandas as pd
import numpy as np
from numpy import median
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set up mysql connector
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  passwd="2514",
  database="immigration"
)

mycursor = mydb.cursor()
```

```python
# Return average salary for certain certified jobs across the nation
select_str = """
select WORKSITE_STATE, count(case_number), round(avg(WAGE_RATE_OF_PAY_FROM),0) 
from h1b_hire_salary 
where CASE_STATUS='Certified' and JOB_TITLE like "%data scientist%" and WAGE_UNIT_OF_PAY='Year'
group by WORKSITE_STATE;
"""

select_str1 = """
select job_title, count(case_number), round(avg(WAGE_RATE_OF_PAY_FROM),0) 
from h1b_hire_salary 
where CASE_STATUS='Certified' and worksite_state='NY' and WAGE_UNIT_OF_PAY='Year'
group by job_title
having count(case_number)>=35
order by avg(WAGE_RATE_OF_PAY_FROM);
"""
```

```python
select_str1 = """
select WORKSITE_STATE, count(case_number), round(avg(WAGE_RATE_OF_PAY_FROM),0) 
from h1b_hire_salary 
where CASE_STATUS='Certified' and JOB_TITLE like '%
"""
select_str2 = """
%' and WAGE_UNIT_OF_PAY='Year'
group by WORKSITE_STATE;
"""

def sy_job_state(job_title):
    job_search_str = select_str1+job_title+select_str2
    print(job_search_str)
    mycursor.execute("""
        select WORKSITE_STATE, count(case_number), round(avg(WAGE_RATE_OF_PAY_FROM),0) 
        from h1b_hire_salary 
        where CASE_STATUS='Certified' and JOB_TITLE like "%"""+"data scientist"+"""%" and WAGE_UNIT_OF_PAY='Year'
        group by WORKSITE_STATE;
        """)
    myresult = mycursor.fetchall()
    datasci_avg_salary = pd.DataFrame(myresult, columns=['State','Counts','Avg_Salary'])
    datasci_avg_salary = datasci_avg_salary.sort_values(by='Avg_Salary', ascending=True)
    print(datasci_avg_salary )
    fig, ax = plt.subplots(figsize=(17,7))
    g = sns.lineplot(x="State",  y="Counts",sort= False, data=datasci_avg_salary, ax=ax, alpha = 1, color = 'r')
    ax2 = ax.twinx()
    ax2.grid(False)
    g = sns.barplot(x="State",  y="Avg_Salary", data=datasci_avg_salary, ax=ax2, alpha=.5)
    ax.set_title(job_title +' Average Salary v.s. Counts', fontsize = 17)
    plt.show()
    print(datasci_avg_salary)

sy_job_state("data scientist")   
```

```python
mycursor.execute(select_str)
myresult = mycursor.fetchall()
datasci_avg_salary = pd.DataFrame(myresult, columns=['State','Counts','Avg_Salary'])
```
